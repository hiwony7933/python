```python
# -*- coding: utf-8 -*-
import os, time
import shutil
 
 
# 지정한 경로의 이미지 파일을 생성한 날짜를 읽어서
# 대상 경로에 폴더 만들어서 이미지를 이동 시킨다.
#검색할 경로
imagepath = '/volume1/admin/phone'
#이동할 대상
targetpath = '/volume1/homes/photo/photo'
 
with open('filecopy_complete.txt', 'r') as fs:
    filelist = fs.readlines()
    fs.close()
 
 
#파일 목록 가져오기
for root, dirs, files in os.walk(imagepath):
    for f in files:
 
        # jpg 인가?
        FilePath = os.path.join(root, f)
        if FilePath.find('JPG') == -1 and FilePath.find('jpg') == -1 :
            continue
    
        Match = False
        for comp_filename in filelist:
            strcomp = comp_filename.replace('\n','')
            strcomp2 = f.replace('\n','')
            
            if strcomp == strcomp2:
                Match = True
                break
                
        # 복사 작업을 했다.
        if Match :
            continue
            
        fs = open('filecopy_complete.txt', 'a')
        fs.write(f+'\n')
        fs.close()
        
        print 'File ADD ' + f
        
        # 파일 생성날짜
        filestat = os.stat( FilePath )
 
        # 접근한 날짜 : st_atime
        # 수정한 날짜 : st_mtime
        # 만든 날짜 : st_ctime
        FolderName = targetpath + '/' + time.strftime('20%y_%m', time.localtime(filestat.st_mtime))
 
        #print time.strftime('%y_%m_%d', time.localtime(filestat.st_ctime))
        
        if not os.path.isdir(FolderName) : 
            os.mkdir(FolderName) 
        
        if not os.path.isfile(FolderName + '/' + f) : 
            #shutil.move (FilePath, FolderName)
            # Move 로 했더니 폰에서 싱크가 계속 일어남
            shutil.copy (FilePath, FolderName)
            print "File Copy == " + FilePath
```
