```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import platform
from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
```

```python
titanic = sns.load_dataset('titanic')
print(titanic.head())
```

```python
#    (5: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

#    (figure 4   )
fig = plt.figure(figsize=(15, 10)) #   
ax1 = fig.add_subplot(2, 2, 1) # 2 2 - 1 
ax2 = fig.add_subplot(2, 2, 2) # 2 2 - 2 
ax3 = fig.add_subplot(2, 2, 3) # 2 2 - 3 
ax4 = fig.add_subplot(2, 2, 4) # 2 2 - 4 

# 1 -   -  
sns.boxplot(x='alive', y='age', data=titanic, ax=ax1)

# 2 -   
#   20~30 ,  20~40     .
sns.boxplot(x='alive', y='age', data=titanic, ax=ax2, hue='sex')

# 3 -   
sns.violinplot(x='alive', y='age', data=titanic, ax=ax3)

# 4 -   
#   30     ,  20~30       .
sns.violinplot(x='alive', y='age', data=titanic, ax=ax4, hue='sex')

ax1.set_title('alive')
ax2.set_title('alive(sex)')
ax3.set_title('alive')
ax4.set_title('alive(sex)')

plt.show()
```
