```python
import platform
from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.xlabel('x') # x  (   )
plt.ylabel('y') # y 
plt.plot(x,y)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]
plt.figure(figsize=(8,6)) #  
plt.plot(x,y,color='green') # : green
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,'bv') 
plt.show() 

# plot() b-   : (b) (-) 

#  : 'r-', 'g-', 'b-'
#  : 'ro', 'go', 'bo'
#   : 'rv', 'gv', 'bv'

# ro  (r) o 
# bv  (b) v 
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

#   

plt.figure(figsize=(8,6))   #  
plt.plot(x, y, color='green', linestyle='dashed')   #   
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

#  : plot() linestyle='dashed' 
#   : plot() marker = 'o' 

plt.figure(figsize=(8,6))
plt.plot (x, y, 
    color='red', 
    linestyle='dashed', 
    marker='o')
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

#  : plot() linestyle='dashed' 
#   : plot() marker = 'o' 
#   : plot() markerfacecolor = 'red'
#   : plot() markersize = 12

plt.figure(figsize=(8,6))
plt.plot(x, y,
    color='green', 
    linestyle='dashed', 
    marker='o', 
    markerfacecolor='red', 
    markersize=9)
plt.show()
```
