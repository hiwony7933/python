```python
# XML 파싱
# http://www.hani.co.kr/rss/ 에서 item태그의 title만 추출

#데이터 가져오기
import requests
resp = requests.get('http://www.hani.co.kr/rss/lead/')
text = resp.text
print(text)
```

```python
import bs4
#트리로 펼쳐내기 
bs = bs4.BeautifulSoup(text, 'lxml-xml')
print(bs)
```

```python
items = bs.find_all('item')
print(items)
```

```python
for i in items : 
    cate = i.find('dc:category')
    date = i.find('pubDate')
    title = i.find('title')
    link = i.find('link')
    print(date.getText(), cate.getText(), title.getText(), link.getText())
```

```python
import pandas as pd
df = pd.DataFrame(columns=['date', 'category','title', 'link'])
for i in items : 
    cate = i.find('dc:category')
    date = i.find('pubDate')
    title = i.find('title')
    link = i.find('link')
    df = df.append(pd.DataFrame([[date, cate, title, link]], columns=['date', 'category','title', 'link']))

df.set_index('date', inplace=True)
print(df)
```

```python
# 파일 저장
df.to_csv(('hani.csv'))
```
